```python
# Catch Rate Task (2022)
players = [{
   'name': 'A.J. Brown',
   'catches': 88,
   'targets': 145
   },
   {
   'name': 'CeeDee Lamb',
   'catches': 107,
   'targets': 156
   },
   {
   'name': 'Justin Jefferson',
   'catches': 128,
   'targets': 184
   },
]

for player in players:
    name = player["name"]
    catches = player["catches"]
    targets = player["targets"]
    catch_rate = catches/targets
    print(name + ' has a catch rate of ' + str(catch_rate))
```
